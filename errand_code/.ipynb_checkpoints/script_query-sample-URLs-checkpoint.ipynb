```python
# Attempt number 1: see if we can Google these, and pull the # of results
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pickle 

shopping_list = ['chicken thighs', 'apples','shampoo', 'chorizo', 'venison', 'lemongrass', 'bacon',
                 'lemongrass','manicure', 'bicycle chain','milk','oranges','rotate tires','vermicelli','fertilizer']

url_df = pickle.load(open( "df_URLs.p", "rb" ) )
url_df.set_index('business_id',inplace=True)

nearby_stores = pickle.load(open( "nearby_stores.p", "rb" ) )

#Open up headless browser to count number of links/hits we get off a query term search on store's URL 
options1 = webdriver.ChromeOptions()
#options1.add_argument('headless')
#options1.add_argument('window-size=1200x600')
# initialize the driver
driver = webdriver.Chrome('./chromedriver', options=options1)
driver.get('https://google.com/');
id_list = nearby_stores['business_id'].tolist();

for i in range(len(id_list)):
    store_results =np.zeros(len(shopping_list))
    b_id = id_list[i]
    print(i,')', b_id)
    tmp_url = (url_df['url'].loc[b_id]).split('/')[2]
    tmp_url = (tmp_url.split('.')[-2])+'.'+(tmp_url.split('.')[-1])
    for j,query_item in enumerate(shopping_list):  
        query1 = '"'+query_item+'"'+' site:'+tmp_url
        time.sleep(0.5)
        search_box = driver.find_element_by_name('q')
        search_box.clear()
        search_box.send_keys(query1)
        search_box.submit();
        time.sleep(1)
        # Parse .html page for "resultsStats"
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        results_summary = soup.find_all('div',id='resultStats')
        # Get the number of results from the results summary
        if len(results_summary)>0:
            tmp1 = results_summary[0].text
            if (tmp1[0].lower() == 'a'): # Drop 'about', if present
                tmp1 = tmp1[6:]
            store_results[j] = float(tmp1[0:tmp1.find(' result')].replace(',',''))


    pickle.dump(store_results, open( 'query_counts/'+b_id+".p", "wb" ) )
```
